```python
# SGD 50 epochs, Adam 50 epochs
# 텐서플로우로 선형회귀 모델링 선형회귀모델을 만들고 ‘5’를 넣었을때 값을 예측.

# x = np.array([2,4,6,8,10])
# y = np.array([20,24,37,40,50])
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([2, 4, 6, 8, 10], dtype=float)
y = np.array([20, 24, 37, 40, 50], dtype=float)

model = Sequential()

# 입력 레이어와 첫 번째 은닉 레이어 추가
model.add(Dense(32, input_dim=1, activation='relu'))

# 출력 레이어 추가
model.add(Dense(1, activation='linear'))  # 선형 활성화 함수

# 모델 컴파일
model.compile(loss='mean_squared_error', optimizer='SGD', metrics=['accuracy'])

# 모델 훈련
model.fit(x, y, epochs=50, batch_size=10)

# 5를 입력으로 넣어 값을 예측
predicted_value = model.predict(np.array([5.0]))
print("입력 값 5에 대한 예측값:", predicted_value[0][0])
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

x = np.array([2, 4, 6, 8, 10], dtype=float)
y = np.array([20, 24, 37, 40, 50], dtype=float)

model = Sequential()

# 입력 레이어와 첫 번째 은닉 레이어 추가
model.add(Dense(32, input_dim=1, activation='relu'))

# 출력 레이어 추가
model.add(Dense(1, activation='linear'))  # 선형 활성화 함수

# 모델 컴파일
model.compile(loss='mean_squared_error', optimizer='Adam', metrics=['accuracy'])

# 모델 훈련
model.fit(x, y, epochs=50, batch_size=10)

# 5를 입력으로 넣어 값을 예측
predicted_value = model.predict(np.array([5.0]))
print("입력 값 5에 대한 예측값:", predicted_value[0][0])
```
