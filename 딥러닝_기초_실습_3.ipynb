```python
import numpy as np
import tensorflow as tf
```

```python
# 데이터
x = np.array([2, 4, 6, 8, 10], dtype=np.float32)
y = np.array([0, 0, 0, 1, 1], dtype=np.float32)
```

```python
# 로지스틱 회귀 모델 생성
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, activation='sigmoid', input_shape=(1,))
])

# 모델 컴파일
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 모델 학습 (데이터양이 적어서 많이하면 오버피팅)
model.fit(x, y, epochs=10, verbose=0)

# '9'를 입력하여 값을 예측
input_value = np.array([9], dtype=np.float32)
predicted_value = model.predict(input_value)

print("예측 결과:", predicted_value[0][0])
```
